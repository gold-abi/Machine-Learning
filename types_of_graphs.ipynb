```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare')
plt.title("Titanic Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.barplot(data=titanic, x='age', y='fare')
plt.title("Titanic Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.lineplot(data=titanic, x='age', y='fare')
plt.title("Titanic Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.boxplot(data=titanic, x='age', y='fare')
plt.title("Titanic Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.violinplot(data=titanic, x='age', y='fare')
plt.title("Titanic Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
import numpy as np
iris=sns.load_dataset('iris')
correlation_matrix=iris.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.jointplot(data=titanic, x='age', y='fare',kind='hex')
plt.title("Titanic Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.violinplot(data=titanic, x='class')
plt.title("passenger class distribution")
plt.xlabel("class")
plt.ylabel("count")
plt.show()
```

```python
sns.countplot(data=titanic, x='class')
plt.title("passenger class distribution")
plt.xlabel("class")
plt.ylabel("count")
plt.show()
```
